```python
#Data Showing the flight Delays Using Heatmaps and Bar charts
flight_data = pd.read_csv("flight_delays.csv",index_col="Month")

flight_data.head()
```

```python
#plotting the data
plt.figure(figsize=(10,6))


#Add title
plt.title("Average Arrival Delay for HA Airlines Flights, by Month")


#bar chart showing average delays of a specific airline
sns.barplot(x=flight_data.index,y=flight_data['HA'])

#Add label to y axis
plt.ylabel("Arrival delay (in minutes)")
```

```python
#HeatMap
#plotting the data
plt.figure(figsize=(14,7))


#Add title
plt.title("Average Arrival Delay for each Airlines Flights, by Month")

#Heatmap showing average delay of each airline by month
sns.heatmap(data=flight_data,annot=True)

#Addlabel
plt.xlabel("Airline")
```

```python
#Scatter pLots -synthetic dataset of insurance charges to see if we understand why customers pay more than others
#Data Showing the flight Delays Using Heatmaps and Bar charts
insurance_data = pd.read_csv("insurance.csv")

insurance_data.head()
```

```python
#simple Scatter plot for Bmi vs charges
sns.scatterplot(x=insurance_data['bmi'],y=insurance_data['charges'])
```

```python
#to check the relationship, we can use regression line
sns.regplot(x=insurance_data['bmi'],y=insurance_data['charges'])
```

```python
#relationship with colour plots for 3 variables using colour coding
sns.scatterplot(x=insurance_data['bmi'],y=insurance_data['charges'],hue=insurance_data['smoker'])
```

```python
#lmplot to create regressionline fornon smokers and smokers
sns.lmplot(x="bmi",y="charges",hue="smoker",data=insurance_data)
```

```python
#Swarmplot categorical scatterplot
sns.swarmplot(x=insurance_data['smoker'],y=insurance_data['charges'])
```

```python
#Histogram and Density plots - 150  flowers and 3 differenct species
flower_data = pd.read_csv("iris.csv",index_col="Id")

flower_data.head()
```

```python
#Histogram for petallengths
sns.distplot(a=flower_data['Petal Length (cm)'],kde=False)
```

```python
#Smoothed Histogram plot with Kernel density estimate
sns.kdeplot(data=flower_data['Petal Length (cm)'],shade=True)
```

```python
#2D KDE Plot
sns.jointplot(x=flower_data['Petal Length (cm)'],y=flower_data['Sepal Width (cm)'],kind="kde")
```

```python
#Plotting to  real-world dataset containing information collected from microscopic images of breast cancer tumors
cancer_b_data = pd.read_csv("cancer_b.csv",index_col="Id")
cancer_m_data = pd.read_csv("cancer_m.csv",index_col="Id")

cancer_b_data.head()
```

```python
#Histogram for benign and Maligant tumors
sns.distplot(a=cancer_b_data['Area (mean)'],label="benign",kde=False)
sns.distplot(a=cancer_m_data['Area (mean)'],label="Maligant",kde=False)


plt.legend()
```

```python
#KDE Plots for different species
sns.kdeplot(data=cancer_b_data['Radius (worst)'],label="benign",shade=True)
sns.kdeplot(data=cancer_m_data['Radius (worst)'],label="Maligant",shade=True)

```
